```
ind = 2

%tensorflow_version 1.x
import gpt_2_simple as gpt2
from datetime import datetime
from google.colab import files

file_names = ["realdonaldtrump_tweets.csv",
              "elonmusk_tweets.csv",
              "kanyewest_tweets.csv",
              "joebiden_tweets.csv"]

sess = gpt2.start_tf_sess()

run = file_names[ind].split("_")[0]

gpt2.finetune(sess,
              dataset=file_names[ind],
              model_name='124M',
              steps=150,
              restore_from='fresh',
              run_name=run,
              save_every=150,
              print_every=10)
              #sample_every=200,
              #save_every=1
              #)
```

```
prefixes = ["EconLab beer is",
            "This beer makes me",
            "I'd never say this about a beer, but",
            "EconLab's beer tastes like",
            "Beer.",
            "This beer will make you want to",
            "This beer is",
            "This beer is just"]
```

```
for prefix in prefixes:
    print()
    gpt2.generate(sess,
                  run_name=run,
                  length=200,
                  temperature=0.7,
                  prefix=prefix,
                  nsamples=10,
                  truncate='<|endoftext|>'
                  )
```
